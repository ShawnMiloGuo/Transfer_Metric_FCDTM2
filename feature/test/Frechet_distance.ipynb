```python
import torch
from scipy.linalg import sqrtm

def covariance_matrix(tensor1, tensor2):
    # 计算均值
    mean1 = torch.mean(tensor1, dim=0)
    mean2 = torch.mean(tensor2, dim=0)

    # 去均值化
    tensor1_centered = tensor1 - mean1
    tensor2_centered = tensor2 - mean2

    # 计算协方差矩阵
    cov_matrix = torch.mm(tensor1_centered.t(), tensor2_centered) / (tensor1_centered.shape[0] - 1)

    return cov_matrix
```

```python
real_features = torch.tensor([[[[1., 2]],

                                [[2, 4]],

                                [[1, 2]],

                                [[3, 4]]]])
fake_features = torch.tensor([[[[1., 2]],

                                [[3, 4]],

                                [[1, 2]],

                                [[3, 4]]]])
real_features_permute = real_features.permute(0, 2, 3, 1).reshape(-1, 4)
print("\nreal_features_permute: \n", real_features_permute)
sigma_real = torch.cov(real_features_permute.t())
print("\nsigma_real: \n", sigma_real)

fake_features_permute = fake_features.permute(0, 2, 3, 1).reshape(-1, 4)
print("\nfake_features_permute: \n", fake_features_permute)
sigma_fake = torch.cov(fake_features_permute.t())
print("\nsigma_fake: \n",sigma_fake)

# print("covariance_matrix(real_features_permute, real_features_permute)\n", covariance_matrix(real_features_permute, real_features_permute))
# print("covariance_matrix(real_features_permute, fake_features_permute)\n", covariance_matrix(real_features_permute, fake_features_permute))

mu_real = torch.mean(real_features_permute, dim=0)
print("\nmu_real: \n", mu_real)
mu_fake = torch.mean(fake_features_permute, dim=0)
print("\nmu_fake: \n", mu_fake)
```

```python
# 
import numpy as np

sigma_dot = torch.mm(sigma_real, sigma_fake)
print("\nsigma_dot: \n", sigma_dot)
# 
# print("\nsigma_real @ sigma_fake \n", sigma_real @ sigma_fake)

sqrt_sigma_dot = sqrtm(sigma_dot)
print("\nsqrt_sigma_dot: \n", sqrt_sigma_dot)
if np.iscomplexobj(sqrt_sigma_dot):
    sqrt_sigma_dot = sqrt_sigma_dot.real

tr_cov = torch.trace(sigma_real + sigma_fake - 2 * sqrt_sigma_dot)
print(tr_cov)
```

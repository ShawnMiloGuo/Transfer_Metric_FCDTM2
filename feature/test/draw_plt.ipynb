```python
import matplotlib.pyplot as plt

# 创建数据点
x = [0, 10]
y1 = [0, 4]  # 第一条直线的y值
y2 = [0, 2]    # 第二条直线的y值

# 绘制第一条直线
plt.plot(x, y1, label='Line 1', color='blue', marker='')

# 绘制第二条直线
plt.plot(x, y2, label='Line 2', color='red', marker='')

# 添加图例
plt.legend()

# 添加标题和标签
plt.title('Two Lines on the Same Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 显示图形
plt.show()
```

```python
import numpy as np
from scipy.linalg import sqrtm

# 示例二维数据矩阵，假设我们要忽略0值
data = np.array([[1., 2, 3, 4],
                 [4, 0, 2, 1],])

# 将0值替换为np.nan
data_with_nan = np.where(data == 0, np.nan, data)

# 创建掩码数组，自动将np.nan视为掩码
masked_data = np.ma.array(data_with_nan, mask=np.isnan(data_with_nan))
print(masked_data)

mean_matrix = np.ma.mean(masked_data, axis=0)
print(mean_matrix)

# 计算掩码数组的协方差矩阵
cov_matrix = np.ma.cov(masked_data, rowvar=False)
print(cov_matrix)

sigma1 = np.ma.filled(cov_matrix, 0)
covmean = sqrtm(sigma1.dot(sigma1))
print(covmean)

print(np.trace(cov_matrix))
```

```python
import numpy as np
from scipy.linalg import sqrtm

# 示例二维数据矩阵，假设我们要忽略0值
data = np.array([[1., 0, 3, 4],
                 [4, 0, 2, 1],])

data2 = np.array([[1., 16, 3, 4],
                    [4, 0, 2, 1],])


# 创建掩码数组，自动将0视为掩码
masked_data = np.ma.array(data, mask=np.where(data == 0.0, True, False))
print(masked_data)

mean_matrix = np.ma.mean(masked_data, axis=0)
mean_matrix = np.ma.filled(mean_matrix, 0)
mean_matrix2 = np.ma.mean(data2, axis=0)
ssdiff = np.sum((mean_matrix - mean_matrix2)**2.0)
print(mean_matrix, mean_matrix2)
print("ssdiff:",ssdiff)

# 计算掩码数组的协方差矩阵
cov_matrix = np.ma.cov(masked_data, rowvar=False)
print(cov_matrix)

sigma1 = np.ma.filled(cov_matrix, 0)
covmean = sqrtm(sigma1.dot(sigma1))
print(covmean)

print(np.trace(cov_matrix))
```

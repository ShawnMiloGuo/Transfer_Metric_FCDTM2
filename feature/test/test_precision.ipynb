```python
import torch
predictions = [0, 0, 0, 1, 1, 1, 1, 1, 1]
labels =      [0, 0, 1, 1, 0, 1, 1, 1, 1]
num_classes = 2
ignore_index = 0
precision = multiclass_precision(torch.tensor(predictions), torch.tensor(labels), num_classes, "none", ignore_index=ignore_index)
recall = multiclass_recall(torch.tensor(predictions), torch.tensor(labels), num_classes, "none", ignore_index=ignore_index)
f1 = multiclass_f1_score(torch.tensor(predictions), torch.tensor(labels), num_classes, "none", ignore_index=ignore_index)
print(precision)
print(recall)
print(f1)
precision = multiclass_precision(torch.tensor(predictions), torch.tensor(labels), num_classes, "micro", ignore_index=ignore_index)
recall = multiclass_recall(torch.tensor(predictions), torch.tensor(labels), num_classes, "micro", ignore_index=ignore_index)
f1 = multiclass_f1_score(torch.tensor(predictions), torch.tensor(labels), num_classes, "micro", ignore_index=ignore_index)
print(precision)
print(recall)
print(f1)
precision = multiclass_precision(torch.tensor(predictions), torch.tensor(labels), num_classes, "macro", ignore_index=ignore_index)
recall = multiclass_recall(torch.tensor(predictions), torch.tensor(labels), num_classes, "macro", ignore_index=ignore_index)
f1 = multiclass_f1_score(torch.tensor(predictions), torch.tensor(labels), num_classes, "macro", ignore_index=ignore_index)
print(precision)
print(recall)
print(f1)
# print((2/4 + 3/4) / 2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

# 设置随机数种子
np.random.seed(3)

# 生成100个随机点
x = np.random.normal(40, 20, 100)

# 设置线性关系的参数
m = 0.01  # 斜率
b = 0.0   # 截距
noise = np.random.normal(0, 0.05, 100)  # 正态分布的噪声

# 生成线性相关的y值
y = m * x + b + noise

# 创建一个方形的绘图区域
plt.figure(figsize=(6, 4))

# 绘制散点图
plt.scatter(x, y, color='blue', alpha=0.5)

# 设置x轴和y轴的取值范围
plt.xlim(-20, 100)
plt.ylim(-0.4, 1.0)


# 设置字体以支持中文
rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 使用黑体
rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
# 添加标题和标签
# plt.title("Linear Relationship")
plt.xlabel("可迁移性度量分数")
plt.ylabel("迁移后精度损失")

# 显示图形
plt.show()
```

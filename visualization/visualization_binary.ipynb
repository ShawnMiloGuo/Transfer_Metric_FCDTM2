```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

# 假设你有三个图像：image, label, prediction
# 这些图像可以是numpy数组，也可以是PIL.Image对象
# 在这个例子中，我们假设它们都是numpy数组

np.random.seed(0)
# image = np.random.randint(0, 256, (256, 256, 3), dtype=np.uint8)
# label = np.random.choice([0, 255], (256, 256, 1))
# prediction = np.random.choice([0, 255], (256, 256, 1))
image = np.array([[[255,0,0],[0,255,0]],
                  [[0,0,255],[255,255,255]]])
label = np.array([[[255],[0]],
                  [[0],[255]]])
prediction = np.array([[[255],[0]],
                  [[0],[255]]])

# 创建一个新的图像
plt.figure()

# 添加第一个子图：原始图像
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Image')
# plt.axis('off')


# 添加第二个子图：标签
plt.subplot(1, 3, 2)
plt.imshow(label, cmap='gray')
plt.title('Label')
# plt.axis('off')
plt.xticks([])
plt.yticks([])


# 添加第三个子图：预测
plt.subplot(1, 3, 3)
# plt.imshow(prediction, cmap='gray')
plt.imshow(np.where(prediction == 255, [50, 0, 255], [0, 0, 0]))
plt.title('Prediction')
# plt.axis('off')
plt.xticks([])
plt.yticks([])


# 显示图像
plt.show()

# 保存图像
# plt.savefig('output.png')
plt.close()
```
